```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('odi_df.csv')
df.head()
```

```python
df.shape
```

```python
score_map={}

for ind,row in df.iterrows():
    a=row['Team1']+' vs '+row['Team2']
    b=row['Team2']+' vs '+row['Team1']
    
    score_map[a]=row['ScoreID']
    score_map[b]=row['ScoreID']

score_map
```

Extracting batting Data to csv

```python
df_1=pd.read_csv('odi_bat_summary.csv')
df_1.head(10)
```

```python
df_1.info()
```

```python
df_1.shape
df_1['Dismissal'].isnull().head(10)
```

```python
df_1['out']=df_1.Dismissal.apply(lambda x: "0" if x is None else "1")
df_1.head(20)
```

```python
df_1.loc[df_1['Dismissal'].isnull(),'out']=0
df_1.head(20)
```

```python
df_1.drop(columns=["Dismissal"],inplace=True)
```

```python
df_1.tail(20)
```

Removing special characters from batsman name

```python
df_1['Batsman']=df_1['Batsman'].apply(lambda x: x.replace('(c)',''))
df_1['Batsman']=df_1['Batsman'].apply(lambda x: x.replace('†',''))
df_1.head(10)
```

```python
df_1['ScoreID']=df_1['Match'].map(score_map)
df_1.tail(10)
```

```python
df_2=pd.read_csv('odi_bowl_fact_summary.csv')
df_2.head()
```

```python
df_2['ScoreID']=df_2['Match'].map(score_map)
df_2.tail(10)
```

```python
df_3=pd.read_csv('player_Details.csv')
df_3.head()
```

```python
df_3.drop(columns=['Unnamed: 7'],inplace=True)
```

```python
df_3.head(15)
```

```python
df_3['player']=df_3['player'].apply(lambda x: x.replace('(c)',''))
df_3['player']=df_3['player'].apply(lambda x: x.replace('(vc)',''))
df_3['player']=df_3['player'].apply(lambda x: x.replace('(wk)',''))
df_3.head(15)
```

```python
df.to_csv("odi_df.csv", index = False)
df_1.to_csv("odi_bat_summary.csv", index = False)
df_2.to_csv("odi_bowl_fact_summary.csv", index = False)
df_3.to_csv("player_Details.csv", index = False)
```
